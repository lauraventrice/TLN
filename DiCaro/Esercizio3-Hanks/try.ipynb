```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
print(spacy.explain("xcomp"))
print(spacy.explain("relcl"))
print(spacy.explain("appos"))
print(spacy.explain("csubjpass"))
print(spacy.explain("dative"))

doc = nlp("One day, at work, he was using an iron rod to handle explosive gun powder.")

displacy.render(doc, style="dep")
```
