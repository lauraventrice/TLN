```python
from concepts import Context

```

```python
# 1. Load data

path_concepts_data = f'resource/fca_data.csv'
concepts_data = Context.fromfile(path_concepts_data, frmat='csv')

print("FCA data loaded ✔")

print(concepts_data)
```

```python
# 2. Queries 

# 2.1 Query common properties of objects or common objects of properties (derivation)
print(concepts_data.intension(['Lion', 'Giraffa']))

print(concepts_data.extension(['Forest', 'Warm-blood']))

# 2.2 Get the closest matching objects-properties pair of objects or properties (formal concepts)

print(concepts_data['Panda', 'Koala'])

print(concepts_data['Viviparous',])
```
